```python
from pathlib import Path
import os

from flair.data import Dictionary
from flair.models import LanguageModel
from flair.trainers.language_model_trainer import LanguageModelTrainer, TextCorpus

from torch.optim.adam import Adam

result_folder = '/home/jupyter/language_model/'
data_folder = '/home/jupyter/data_ner/'
corpus_folder = os.path.join(data_folder, 'char_corpus')
# load the default character dictionary
dictionary: Dictionary = Dictionary.load(os.path.join(data_folder, 'ara_char_mappings'))


forward_and_backward = [(True, 'wiki-forward'), (False, 'wiki-backward')]
for is_forward_lm, fname in forward_and_backward:
    # are you training a forward or backward LM?
    # get your corpus, process forward and at the character level
    corpus = TextCorpus(Path(corpus_folder),
                        dictionary,
                        is_forward_lm,
                        character_level=True)

    # instantiate your language model, set hidden size and number of layers
    language_model = LanguageModel(dictionary,
                                   is_forward_lm,
                                   hidden_size=128,
                                   nlayers=1)

    # train your language model
    trainer = LanguageModelTrainer(language_model, corpus, optimizer=Adam)
    trainer.train(os.path.join(result_folder, fname),
              sequence_length=100,
              mini_batch_size=1000,
              max_epochs=500)
```
